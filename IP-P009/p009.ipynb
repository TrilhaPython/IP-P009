1. Carregar agora os dados dos dois datasets utilizando um DateTimeIndex ;

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataframe2003 = pd.read_csv('./datasets/INMET_CO_MT_A901_CUIABA_01-01-2003_A_31-12-2003.CSV', sep=';', encoding='ISO-8859-1' , decimal = ',')
dataframe2013 = pd.read_csv('./datasets/INMET_CO_MT_A901_CUIABA_01-01-2013_A_31-12-2013.CSV', sep=';', encoding='ISO-8859-1', decimal = ',')
dataframe2023 = pd.read_csv('./datasets/INMET_CO_MT_A901_CUIABA_01-01-2023_A_31-12-2023.CSV', sep=';', encoding='ISO-8859-1', decimal = ',')
```

```python
#Renomeando coluna do datasets
dataframe2023.rename(columns={'Hora UTC':'HORA (UTC)', 'RADIACAO GLOBAL (Kj/mý)': 'RADIACAO GLOBAL (KJ/mý)'}, inplace=True)
dataframe2003.rename(columns={'PRECIPITAÇÃO TOTAL, HORÁRIO (MM)': 'PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'},inplace=True)



#Convertendo coluna DATA para datetime
dataframe2003['DATA'] = pd.to_datetime(dataframe2003['DATA'], format='%d/%m/%Y', errors='coerce').dt.strftime('%Y/%m/%d')
dataframe2013['DATA'] = pd.to_datetime(dataframe2013['DATA'], format='%d/%m/%Y', errors='coerce').dt.strftime('%Y/%m/%d')
dataframe2023['DATA'] = pd.to_datetime(dataframe2023['DATA'], format='%d/%m/%Y', errors='coerce').dt.strftime('%Y/%m/%d')

#Tratando HORA (UTC)
dataframe2023['HORA (UTC)'] = dataframe2023['HORA (UTC)'].str.rstrip(' UTC').str.slice_replace(2, 2, ':')

#Juntando datasets
df_concat = pd.concat([dataframe2003, dataframe2013, dataframe2023])
df_concat
```

```python
df_concat.replace(-9999.0, np.nan, inplace=True)
```

```python
df_concat
```

```python
#Definindo coluna data como indice
data = df_concat["DATA"].values
index = pd.DatetimeIndex(data, name="Data")
df_concat.drop("DATA", axis=1, inplace=True)
df_concat.set_index(index, inplace=True)
df_concat
```

```python
df_concat.columns.tolist()
colunas = [
    ('HORA (UTC)',''),
 ('PRECIPITAÇÃO','PRECIPITAÇÃO TOTAL, HORÁRIO (mm)'),
 ('PRESSAO','PRESSAO ATMOSFERICA AO NIVEL DA ESTACAO, HORARIA (mB)'),
 ('PRESSAO','PRESSÇO ATMOSFERICA MAX.NA HORA ANT. (AUT) (mB)'),
 ('PRESSAO','PRESSÇO ATMOSFERICA MIN. NA HORA ANT. (AUT) (mB)'),
 ('RADIACAO','RADIACAO GLOBAL (KJ/mý)'),
 ('TEMPERATURA','TEMPERATURA DO AR - BULBO SECO, HORARIA (øC)'),
 ('TEMPERATURA','TEMPERATURA DO PONTO DE ORVALHO (øC)'),
 ('TEMPERATURA','TEMPERATURA MµXIMA NA HORA ANT. (AUT) (øC)'),
 ('TEMPERATURA','TEMPERATURA MÖNIMA NA HORA ANT. (AUT) (øC)'),
 ('TEMPERATURA','TEMPERATURA ORVALHO MAX. NA HORA ANT. (AUT) (øC)'),
 ('TEMPERATURA','TEMPERATURA ORVALHO MIN. NA HORA ANT. (AUT) (øC)'),
 ('UMIDADE','UMIDADE REL. MAX. NA HORA ANT. (AUT) (%)'),
 ('UMIDADE','UMIDADE REL. MIN. NA HORA ANT. (AUT) (%)'),
 ('UMIDADE','UMIDADE RELATIVA DO AR, HORARIA (%)'),
 ('VENTO','VENTO, DIRE\x80ÇO HORARIA (gr) (ø (gr))'),
 ('VENTO','VENTO, RAJADA MAXIMA (m/s)'),
 ('VENTO','VENTO, VELOCIDADE HORARIA (m/s)')]

df_concat.columns = pd.MultiIndex.from_tuples(colunas)
```

### Exercício 1: Tratando séries temporais indexadas por tempo

##### 1. Analisando as colunas relacionadas e precipitações e temperaturas, mostre num subplot a evolução das temperaturas médias e acumulado de precipitações nos tres anos.

```python
anos = [2003, 2013, 2023]
df_filtrado = df_concat[df_concat.index.get_level_values("Data").year.isin(anos)]
```

```python
media_mensal = df_filtrado.groupby([df_filtrado.index.get_level_values(0).year, df_filtrado.index.get_level_values(0).month])['PRECIPITAÇÃO'].mean()
```

```python
# Plotando o gráfico
fig, ax = plt.subplots(figsize=(10, 6))
for ano in anos:
    media_mensal.loc[ano].plot(ax=ax, label=str(ano))
    # Adicionando título para cada linha
    ax.lines[-1].set_label(str(ano))

ax.set_xticks(range(1, 13))
ax.set_xticklabels(['Jan', 'Fev', 'Mar', 'Abr', 'Mai', 'Jun', 'Jul', 'Ago', 'Set', 'Out', 'Nov', 'Dez'])
ax.set_xlabel('Mês')
ax.set_ylabel('Média de Precipitação (mm)')
ax.set_title('Média de Precipitação por Mês (2003, 2013, 2023)')
ax.legend(title='Ano', loc='best')
```
